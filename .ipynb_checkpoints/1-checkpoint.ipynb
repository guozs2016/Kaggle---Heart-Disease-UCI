# Heart Disease Neural Network Classifier

Given the data of patients with differnt medical features, we can use a neural network to make prediction.

This is a project of testing building neural network on tensorflow and keras.The aim is to train a model to predict whether a patient has heart disease.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense
```

First we read the data, check the size and take a glance.

```python
data = pd.read_csv('data/heart.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data.dtypes
```

According to the information of data, most of variables are catagorial. We will then process those features with pd.get_dummies()

```python
catagorialList = ['sex','cp','fbs','restecg','exang','ca','thal']
for item in catagorialList:
    data[item] = data[item].astype('object')
```

```python
data.dtypes
```

```python
data = pd.get_dummies(data, drop_first=True)
```

```python
data.head()
```

We will normalize the input features to help neural network train faster.

The factor we use in normalization will be kept to make predictions.

```python
y = data['target'].values
y = y.reshape(y.shape[0],1)
x = data.drop(['target'],axis=1)
minx = np.min(x)
maxx = np.max(x)
x = (x - minx) / (maxx - minx)
x.head()
```

Since the dataset is not large, we will use a 80-20 train-test ratio.

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state=0)
```

In keras, It is quite convenient to setup a neural network model.We will set 21 neurons as input layer, 10 neurons in layer[1], 5 neurons in layer[2] and an output layer

```python
model = Sequential()
model.add(Dense(12, input_dim=21, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=300, batch_size=x_train.shape[0])
```

```python
scores = model.evaluate(x_test, y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
